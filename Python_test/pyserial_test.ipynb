```python
import serial
```

```python
def readserial(comport, baudrate):

    ser = serial.Serial(comport, baudrate, timeout=0.1)         # 1/timeout is the frequency at which the port is read

    while True:
        data = ser.readline().decode().strip()
        if data:
            print(data)



readserial('/dev/cu.usbmodem111659401', 115200)
```

```python
# Importing Libraries 
import serial 
import time 
import keyboard

arduino = serial.Serial(port='/dev/cu.usbmodem111659401', baudrate=9600, timeout=.1)

def write():
	arduino.write(b"RSEN 0\n") 
	time.sleep(0.05) 

write()

while True: 
	id_counter = 0	#num = input("Enter a number: ") # Taking input from user 
	data = arduino.readline().decode().strip()
	time.sleep(0.05) 


	if data != "":
		print(data) # printing the value 

		if keyboard.is_pressed("q"):
        # ESC pressed
			print("Escape hit, closing...")
			break
			
		elif keyboard.is_pressed("Enter"):
        # SPACE pressed
			with open("sensor_data.csv", "w+") as file:
				id_sensor_data= id_counter + data
				file.write(id_sensor_data)
				id_counter += 1
```
