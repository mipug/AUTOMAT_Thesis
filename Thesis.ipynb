```python

def convert_inductance_subtract_baseline(df):
    id_list = ['B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4']
    for id in id_list:
        df[id] = df[id].apply(lambda raw: ((1000000 * ( ( 2**25 ) / (math.pi * raw))**2)))

    baseline = df[df['Weight'] == 0]
    baseline_mean = baseline.mean().iloc[2::]

    for id in id_list:
        df[id] = df[id].apply(lambda raw: raw - baseline_mean[id])
    
    return df

def plot(df=None, sensor=None, weight=None, measure=None, single_sensor=None): # select data for chosen sensor, weight & measurement type (std, mean)
    df2 = df.copy()
    if single_sensor == None:
        test = df2[(df2.Sensor_id == sensor) & (df2.Weight == weight)]
    else: 
        test = df2[(df2.Sensor_id == single_sensor) & (df2.Weight == weight)]
    plottest = test.groupby('New_Index').agg({id: ['mean','std'] for id in id_list})
    if measure == 'std':
        return plottest[(sensor, 'std')]
    else: 
        return plottest[(sensor, 'mean')]

    
"""def create_figure(df, sensor = None, weight = None, measure = None, weights = None, colors = None, title = None):
    x = np.array(np.arange(0,20, 0.5))
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(111)
    if weights != None:
        for weight, color in zip(weights, colors):
            line_=ax.errorbar(x, plot(df, sensor, weight, 'mean'), yerr=plot(df, sensor, weight, 'std'), ls = '--', color=color, label=weight)
    elif sensor == None: 
        for sensor in id_list:
            line=ax.errorbar(x, plot(df, sensor, weight, 'mean'), yerr=plot(df, sensor, weight, 'std'), ls = '--', label=sensor)
    else:
        for sensors in id_list:
            line=ax.errorbar(x, plot(df=df, sensor=sensors, weight=weight, measure='mean', single_sensor=sensor), yerr=plot(df=df, sensor=sensors, weight=weight, measure='std', single_sensor=sensor), ls = '--', label=sensors)

    plt.title(title, fontsize=25) 
    plt.legend()
    ax.set_xlabel('Time in Seconds', fontsize = 18)
    ax.set_ylabel('Raw Output', fontsize = 18)
    plt.show()"""


import numpy as np
import matplotlib.pyplot as plt

def create_figure(df, sensor=None, weight=None, measure=None, weights=None, colors=None, title=None, start_value_zero=False):
    x = np.arange(0, 20, 0.5)
    fig = plt.figure(figsize=(15, 8))
    ax = fig.add_subplot(111)
    
    if weights is not None:
        for weight, color in zip(weights, colors):
            y_values = plot(df, sensor, weight, 'mean')
            if start_value_zero:
                baseline = np.mean(y_values[:3])  # Calculate the average of the first three values
                y_values -= baseline  # Baseline adjustment
            line_ = ax.errorbar(x, y_values, yerr=plot(df, sensor, weight, 'std'), ls='--', color=color, label=weight)
    elif sensor is None: 
        for sensor in id_list:
            y_values = plot(df, sensor, weight, 'mean')
            if start_value_zero:
                baseline = np.mean(y_values[:3])  # Calculate the average of the first three values
                y_values -= baseline  # Baseline adjustment
            line = ax.errorbar(x, y_values, yerr=plot(df, sensor, weight, 'std'), ls='--', label=sensor)
    else:
        for sensors in id_list:
            y_values = plot(df=df, sensor=sensors, weight=weight, measure='mean', single_sensor=sensor)
            if start_value_zero:
                baseline = np.mean(y_values[:3])  # Calculate the average of the first three values
                y_values -= baseline  # Baseline adjustment
            line = ax.errorbar(x, y_values, yerr=plot(df=df, sensor=sensors, weight=weight, measure='std', single_sensor=sensor), ls='--', label=sensors)

    # Calculate the percentage difference from the lowest first value to the largest first value
    first_values = [line.get_ydata()[0] for line in ax.get_lines()]
    lowest_value = min(first_values)
    highest_value = max(first_values)
    percentage_difference = ((highest_value - lowest_value) / lowest_value) * 100

    
    plt.title(title, fontsize=25) 
    plt.legend()
    ax.set_ylabel('Raw Output', fontsize = 18)
    ax.set_xlabel('Time in Seconds', fontsize=18)
    print(lowest_value, highest_value, percentage_difference, highest_value-lowest_value)

```

```python
def import_baselines(experiment_id):
    #loop that imports files
    tetrisbase_file_list = ['tetrisL_baseline', 'tetrisZ_baseline', 'tetrisI_baseline', 'tetrisO_baseline', 'tetrisT_baseline']
    tetris_columns = ['Experiment_id', 'Measurement_id', 'Tetris_id', 'Timestamp', 'B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4', 'MISTAKE']
    df_tetris = pd.DataFrame(columns = tetris_columns)

    for file in tetrisbase_file_list:
        df_temp = pd.read_csv(f'/data/notebook_files/tetris_data/experiment_individualbaseline_1403/{file}.csv', header=None)
        df_temp.columns = tetris_columns
        df_tetris = pd.concat([df_tetris, df_temp], axis=0)
    
    df_tetris.columns = tetris_columns
    df_tetris = df_tetris.drop(['MISTAKE'], axis=1)
    df_tetris = df_tetris[df_tetris['Experiment_id']==experiment_id]
    return df_tetris

#---------------------------------------------------------------------------------

def import_tetris_data(experiment_id):
    tetris_file_list = ['tetris_L', 'tetris_Z', 'tetris_I', 'tetris_O', 'tetris_T']
    tetris_columns = ['Experiment_id', 'Measurement_id', 'Tetris_id', 'Weight', 'Timestamp', 'B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4', 'MISTAKE']
    df_tetris = pd.DataFrame(columns = tetris_columns)

    for file in tetris_file_list:
        df_temp = pd.read_csv(f'/data/notebook_files/tetris_data/experiment_individualbaseline_1403/{file}.csv', header=None)
        df_temp.columns = tetris_columns
        df_tetris = pd.concat([df_tetris, df_temp], axis=0)

    df_tetris.columns = tetris_columns
    df_tetris = df_tetris.drop(['MISTAKE'], axis=1)
    df_tetris = df_tetris[df_tetris['Experiment_id']==experiment_id]
    return df_tetris

#---------------------------------------------------------------------------------

def normalize_data(df_tetris, min_norm, max_norm):
    #Find min value
    #Substracting the mean baseline value from the baseline to find max-values
    df = df_tetris.copy()
    for id in id_list:
        df[id] = df[id].apply(lambda row: (row - max_norm[id]) / (min_norm[id] - max_norm[id]))
    return df
#---------------------------------------------------------------------------------

def subtract_baseline(df_tetris, df_base, mean=True):
    
    id_list = ['B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4']
    df_base_grouped = df_base.groupby(['Experiment_id', 'Measurement_id', 'Tetris_id']).agg({id: ['mean'] for id in id_list})
    df_base_grouped = df_base_grouped.droplevel(1, axis=1)
    
    # group df_tetris on measurement-id, take mean and substract baseline
    if mean:
        df_tetris_grouped = df_tetris.groupby(['Experiment_id', 'Measurement_id', 'Tetris_id']).agg({id: ['mean'] for id in id_list})
        df_tetris_grouped = df_tetris_grouped.droplevel(1, axis=1)
        df = df_tetris_grouped.sub(df_base_grouped)
        return df

    # do not group measurement_id
    else:
        id_list_x = ['B2_x', 'B1_x', 'A2_x', 'A1_x', 'B3_x', 'B0_x', 'A3_x', 'A0_x', 'B6_x', 'B5_x', 'A6_x', 'A5_x', 'B7_x', 'B4_x', 'A7_x', 'A4_x']
        # Merge df_tetris and df_base using merge operation
        merged_df = pd.merge(df_tetris,df_base_grouped, how='left', on=['Experiment_id', 'Measurement_id', 'Tetris_id'])
        for id in id_list:
            merged_df[id] = merged_df[f'{id}_x'] - merged_df[f'{id}_y']

        return merged_df
```

```python
def TestModelOnExperimentData(model, experiment_data, titlee = "Confusion Matrix"):
    model_name = [name for name, value in globals().items() if value is model]
    data_name = [name for name, value in globals().items() if value is experiment_data]
    print(f'Testing model: {model_name[0]}\non dataset: {data_name[0]}')
    # expected to be bad because we took baseline every time in experiment 2
    X_test = experiment_data[id_list]
    y_test = experiment_data["Tetris_id"]

    labels = np.array(['I', 'L', 'O', 'T', 'Z'])
    #print(labels)

    y_one_hot = np.zeros((y_test.shape[0], len(labels)))
    for i, label in enumerate(y_test):
        y_one_hot[i, np.where(labels == label)[0][0]] = 1
        #print(y_one_hot1[i], label)

    predictions = model.predict(X_test)

    correct = 0
    predicted_labels = []
    true_labels = [] 
    for arr, y in zip(predictions, y_one_hot):
        max_index = np.argmax(arr)
        predicted_labels.append(labels[max_index])
        max_yindex = np.argmax(y)
        true_labels.append(labels[max_yindex])
        if max_index == max_yindex:
            correct += 1
        #print('pred: ', max_index, labels[max_index], 'act: ', max_yindex, labels[max_yindex])
    print("Accuracy", correct/len(y_test))
    # Compute confusion matrix
    cm = confusion_matrix(true_labels, predicted_labels)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title(titlee, fontsize = 20)
    plt.show()

def TestModelOnExperimentDataReshaped(model, experiment_data):
    model_name = [name for name, value in globals().items() if value is model]
    data_name = [name for name, value in globals().items() if value is experiment_data]
    print(f'Testing model: {model_name[0]}\non dataset: {data_name[0]}')
    # expected to be bad because we took baseline every time in experiment 2
    X_test = experiment_data[id_list]
    y_test = experiment_data["Tetris_id"]

    X_test = np.array(X_test)
    X_test = X_test.reshape(-1, 4, 4, 1)

    labels = np.array(['I', 'L', 'O', 'T', 'Z'])
    #print(labels)

    # y_one_hot = np.zeros((y_test.shape[0], len(labels)))
    # for i, label in enumerate(y_test):
    #     y_one_hot[i, np.where(labels == label)[0][0]] = 1
        #print(y_one_hot1[i], label)

    y_encoded = label_encoder.fit_transform(y_test)
    y_one_hot = to_categorical(y_encoded)

    predictions = model.predict(X_test)

    correct = 0
    predicted_labels = []
    true_labels = [] 
    for arr, y in zip(predictions, y_one_hot):
        max_index = np.argmax(arr)
        predicted_labels.append(labels[max_index])
        max_yindex = np.argmax(y)
        true_labels.append(labels[max_yindex])
        if max_index == max_yindex:
            correct += 1
        #print('pred: ', max_index, labels[max_index], 'act: ', max_yindex, labels[max_yindex])
    print("Accuracy", correct/len(y_test))

    print("predicted", predicted_labels, "true", true_labels)
    # Compute confusion matrix
    cm = confusion_matrix(true_labels, predicted_labels)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False, xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()

from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python

def build_model(hp):
    dnn_layers_ss = [1, 2, 3]
    dnn_units_min, dnn_units_max = 16, 256
    dropout_ss = [0.1, 0.2]
    lr_min, lr_max = 1e-4, 1e-2
    
    learning_rate = hp.Float('learning_rate', min_value=lr_min, max_value=lr_max, sampling='log')
    
    inputs = tf.keras.Input(shape=(16,))
    dense = inputs 

    for layer_i in range(hp.Choice("n_layers", dnn_layers_ss)):
        dnn_units = hp.Int(f"{layer_i}_units", min_value=dnn_units_min, max_value=dnn_units_max, step=2, sampling='log')
        dense = tf.keras.layers.Dense(units=dnn_units, activation='relu')(dense)
        dropout_rate = hp.Float('dropout_rate', min_value=0.1, max_value=0.5, step=0.1)
        if hp.Boolean("dropout"):
            #dropout_rate = hp.Choice("dropout_rate", values=dropout_ss)  # Define the dropout_rate hyperparameter
            
    
            dense = tf.keras.layers.Dropout(rate=dropout_rate)(dense)
    
    outputs = tf.keras.layers.Dense(units=5, activation='softmax')(dense)
    model = tf.keras.Model(inputs=inputs, outputs=outputs)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss="categorical_crossentropy",
        metrics=["accuracy"],
    )
    return model
```

```python
id_list = ['B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4']


#---------------------------------------------------------------------------------

def convert_inductance(df):
    id_list = ['B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4']
    for id in id_list:
        df[id] = df[id].apply(lambda raw: ((1000000 * ( ( 2**25 ) / (math.pi * raw))**2)))
    return df
```

```python
def import_baselines(experiment_id):
    #loop that imports files
    chickenbase_file_list = ['chicken_b_baseline', 'chicken_c_baseline', 'chicken_d_baseline']
    chicken_columns = ['Experiment_id', 'Measurement_id', 'chicken_id', 'Timestamp', 'B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4', 'MISTAKE']
    df_chicken = pd.DataFrame(columns = chicken_columns)

    for file in chickenbase_file_list:
        df_temp = pd.read_csv(f'/data/notebook_files/chicken_data/{file}.csv', header=None)
        df_temp.columns = chicken_columns
        df_chicken = pd.concat([df_chicken, df_temp], axis=0)
    
    df_chicken.columns = chicken_columns
    df_chicken = df_chicken.drop(['MISTAKE'], axis=1)
    df_chicken = df_chicken[df_chicken['Experiment_id']==experiment_id]
    return df_chicken

#---------------------------------------------------------------------------------

def import_chicken_data(experiment_id):
    chicken_file_list = ['chicken_b', 'chicken_c', 'chicken_d']
    chicken_columns = ['Experiment_id', 'Measurement_id', 'chicken_id', 'Weight', 'Timestamp', 'B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4', 'MISTAKE']
    df_chicken = pd.DataFrame(columns = chicken_columns)

    for file in chicken_file_list:
        df_temp = pd.read_csv(f'/data/notebook_files/chicken_data/{file}.csv', header=None)
        df_temp.columns = chicken_columns
        df_chicken = pd.concat([df_chicken, df_temp], axis=0)

    df_chicken.columns = chicken_columns
    df_chicken = df_chicken.drop(['MISTAKE'], axis=1)
    df_chicken = df_chicken[df_chicken['Experiment_id']==experiment_id]
    return df_chicken

#---------------------------------------------------------------------------------

def subtract_baseline(df_chicken, df_base, mean=True):
    
    id_list = ['B2', 'B1', 'A2', 'A1', 'B3', 'B0', 'A3', 'A0', 'B6', 'B5', 'A6', 'A5', 'B7', 'B4', 'A7', 'A4']
    df_base_grouped = df_base.groupby(['Experiment_id', 'Measurement_id', 'chicken_id']).agg({id: ['mean'] for id in id_list})
    df_base_grouped = df_base_grouped.droplevel(1, axis=1)
    
    # group df_chicken on measurement-id, take mean and substract baseline
    if mean:
        df_chicken_grouped = df_chicken.groupby(['Experiment_id', 'Measurement_id', 'chicken_id']).agg({id: ['mean'] for id in id_list})
        df_chicken_grouped = df_chicken_grouped.droplevel(1, axis=1)
        df = df_chicken_grouped.sub(df_base_grouped)
        return df

    # do not group measurement_id
    else:
        id_list_x = ['B2_x', 'B1_x', 'A2_x', 'A1_x', 'B3_x', 'B0_x', 'A3_x', 'A0_x', 'B6_x', 'B5_x', 'A6_x', 'A5_x', 'B7_x', 'B4_x', 'A7_x', 'A4_x']
        # Merge df_chicken and df_base using merge operation
        merged_df = pd.merge(df_chicken,df_base_grouped, how='left', on=['Experiment_id', 'Measurement_id', 'chicken_id'])
        for id in id_list:
            merged_df[id] = merged_df[f'{id}_x'] - merged_df[f'{id}_y']

        return merged_df
```
